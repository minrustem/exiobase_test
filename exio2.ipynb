```python
import pymrio
import pandas as pd
import os

# Load core IO system
exio = pymrio.parse_exiobase3(
    path='D:/Programming/databases/IOT_2022_pxp'
)

# Just load the physical data - NO calc_system needed for exploration!
data_path = 'D:/Programming/databases/IOT_2022_pxp'

# Load air emissions (physical quantities)
F_air = pd.read_csv(
    os.path.join(data_path, 'air_emissions', 'F.txt'),
    sep='\t',
    index_col=[0, 1],
    header=[0, 1]
)

F_air_Y = pd.read_csv(
    os.path.join(data_path, 'air_emissions', 'F_Y.txt'),
    sep='\t',
    index_col=[0, 1],
    header=[0, 1]
)

print("=== PHYSICAL AIR EMISSIONS ===")
print(f"F matrix (industry emissions): {F_air.shape}")
print(f"F_Y matrix (household emissions): {F_air_Y.shape}")

# Explore physical flows
print("\n1. Available emission types:")
emission_types = F_air.index.get_level_values(1).unique()
print(f"Number of emission types: {len(emission_types)}")
print(emission_types[:20])

# 2. Total physical emissions by type
print("\n2. Total emissions by type (summed across all sectors):")
total_by_type = F_air.sum(axis=1)
print(total_by_type.sort_values(ascending=False).head(10))

# 3. Physical emissions by country
print("\n3. Total emissions by country:")
total_by_country = F_air.groupby(level=0, axis=1).sum().sum(axis=0)
print(total_by_country.sort_values(ascending=False).head(10))
```
